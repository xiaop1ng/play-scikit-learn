```python
from sklearn import linear_model as lr
# 示例：y=2x
reg = lr.LinearRegression()
X = [[-2],[-1], [0], [1], [2]]
y = [-4,-2,0,2,4]
# 输入训练数据
reg.fit(X, y)
# 拟合完毕后，查看模型系数: k => 2
print(reg.coef_)
```

```python
# 使用模型来预测 x=100 时 y 的值
reg.predict([[100]])
```

```python
# 模型评估
from sklearn import metrics

y_pred = reg.predict(X)
print(y_pred)
# 均方误差 MSR
print(metrics.mean_squared_error(y, y_pred))
```

```python
# 可视化
import matplotlib.pyplot as plt
# 原始数据，散点图
plt.plot(X, y, 'ro')
# 模型，线性图
plt.plot(X, y_pred)
plt.title('x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from sklearn import linear_model as lr
# 示例：y=2x+1
reg2 = lr.LinearRegression()
X2 = [[-2],[-1], [0], [1], [2]]
y2 = [-3,-1,1,3,5]
# 输入训练数据
reg2.fit(X2, y2)
# 拟合完毕后，查看模型系数 y=kx+b: k=2, b=1 
print(reg2.coef_)
print(reg2.intercept_)

y_pred2 = reg2.predict(X2)
print(reg2.predict([[100]]))

# 可视化
import matplotlib.pyplot as plt
# 原始数据，散点图
plt.plot(X2, y2, 'ro')
# 模型，线性图
plt.plot(X2, y_pred2)
plt.title('x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
